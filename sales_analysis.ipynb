## Basic Analysis

### Total sales?

```python
df['Sales'].sum()
```

### City-wise total sales?

```python
df.groupby('City')['Sales'].sum()
```

### product-wise total sales?

```python
df.groupby('Product')['Sales'].sum()
```

### Which product has highest sales?

```python
df.groupby('Product')['Sales'].sum().sort_values(ascending=False).head(1)
```

### customer-wise total spending?

```python
df.groupby('Customer')['Sales'].sum()
```

### which city has highest sales?

```python
df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(1)
```

### city-wise average sales?

```python
df.groupby('City')['Sales'].mean()
```

### which is the most frequent product?

```python
df.value_counts(['Product'])
```

### TOP 2 cities by total sales?

```python
df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(2)
```

### Which customer spend the most?

```python
df.groupby('Customer')['Sales'].sum().sort_values(ascending=False).head(1)
```

### filter records whose sales>200?

```python
df[df['Sales']>200]
```

```python
import pandas as pd
data = { 
    'Date': ['2024-01-01','2024-01-02','2024-01-03','2024-01-04','2024-01-05'], 
    'City': ['Pune','Delhi','Pune','Mumbai','Delhi'], 
    'Product': ['A','B','A','C','B'], 
    'Sales': [100,200,150,300,250], 
    'Customer': ['C1','C2','C1','C3','C2']
 }

df = pd.DataFrame(data) 
```

```python
df
```

## Visualization

```python
import matplotlib.pyplot as plt
```

### City-Wise Sales

```python
df.groupby('City')['Sales'].sum().plot(kind='bar')
plt.title('City-wise Sales')
plt.xlabel('City')
plt.ylabel('Sales')
plt.show()
```

### Product_Wise Sales

```python
df.groupby('Product')['Sales'].sum().plot(kind='bar')
plt.title('Product-wise Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

### Sales Distribution

```python
df['Sales'].plot(kind='hist')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.show()
```
